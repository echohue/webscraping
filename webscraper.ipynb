```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
```

```python
url = 'https://www.amazon.in/s?i=fashion&srs=50916365031&rh=n%3A50916365031&s=popularity-rank&fs=true&qid=1750434241&xpid=vyLLd_rZqRbTQ&ref=sr_pg_1'
options = Options()
options.add_argument('--headless')  
options.add_argument('--disable-gpu')
options.add_argument('--no-sandbox')

driver = webdriver.Chrome(options = options)
driver.get(url)



```

```python
soup = BeautifulSoup(driver.page_source,'html.parser')
```

```python
soup_card = soup.find_all('div',class_="sg-col-4-of-4 sg-col-4-of-24 sg-col-4-of-12 s-result-item s-asin sg-col-4-of-16 sg-col s-widget-spacing-small sg-col-4-of-8 sg-col-4-of-20")
```

```python
len(soup_card)
```

```python
product_list = []
brand_list = []
price_list = []
rating_list = []
```

```python
for card in soup_card:
    product_name_div = card.find('h2',class_='a-size-base-plus a-spacing-none a-color-base a-text-normal')
    product_name = product_name_div.find('span').text.strip() if product_name_div else None
    brand = card.find('span',class_='a-size-base-plus a-color-base').text.strip()
    price_div = card.find('span',class_='a-offscreen')
    price = price_div.text.strip() if price_div else None
    rating = card.find('span',class_='a-icon-alt').text.strip()
    product_list.append(product_name)
    brand_list.append(brand)
    price_list.append(price)
    rating_list.append(rating)
    
```

```python
url_flipkart = 'https://www.flipkart.com/search?q=fashion&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1'
```

```python
driver3 = webdriver.Chrome(options = options)
```

```python
driver3.get(url_flipkart)
```

```python
soup3 = BeautifulSoup(driver3.page_source,'html.parser')
```

```python
soup_card3 = soup3.find_all('div',class_='_1sdMkc LFEi7Z')
```

```python
len(soup_card3)
```

```python
link_list = []
for card in soup_card3:
    product_name_div= card.find('a',class_='WKTcLC')
    
    
    product_name = product_name_div.text.strip() if product_name_div else None
    brand_div = card.find('div',class_='syl9yP')
    brand = brand_div.text.strip() if brand_div else None
    price_div = card.find('div',class_='Nx9bqj')
    price = price_div.text.strip() if price_div else None
    links = card.find('a',class_ = 'rPDeLR')['href']
    link_list.append(links)
    product_list.append(product_name)
    brand_list.append(brand)
    price_list.append(price)
    
```

```python
for i in link_list:
    
    driver3.get('https://www.flipkart.com'+i)
    
    soup = BeautifulSoup(driver3.page_source,'html.parser')
    rating_div = soup.find('span',class_='Y1HWO0')
    rating = rating_div.text.strip() if rating_div else None
    rating_list.append(rating)
    
```

```python
ecommerce_site = ['Amazon'] * 57 + ['Flipkart'] * 40
```

```python
df = pd.DataFrame({
    'Product Name':product_list,
    'Brand Name':brand_list,
    'Price':price_list,
    'Rating':rating_list,
    'Ecommerce Site':ecommerce_site
    
})
```

```python
df['Price'] = df['Price'].str.replace('₹', '').str.replace(',','').astype(float)
```

```python
df['Rating'] = df['Rating'].str.extract(r'(\d+\.\d+)').astype(float)
```

```python
category_map = {
    'footwear': ['shoe', 'sandal', 'sneaker', 'flip', 'loafer'],
    'bags': ['bag', 'backpack', 'tote', 'satchel', 'handbag', 'luggage'],
    'clothes': ['saree', 'kurta', 'shirt', 't-shirt', 'blazer', 'sweatshirt', 'jogger', 'pant', 'jeans', 'dress', 'top','kurti','sweater'],
    'accessories': ['watch', 'sunglass', 'earring', 'ring', 'bracelet', 'necklace', 'wallet', 'belt','jewellery']
}
```

```python
keyword_to_category={}
for broad_cat, small_cat in category_map.items():
    for small in small_cat:
        keyword_to_category[small] = broad_cat.capitalize()
    
```

```python
pattern = r'(?i)(' + '|'.join(keyword_to_category.keys()) + r')'
pattern
```

```python
df['Category'] = df['Product Name'].str.extract(pattern,expand = False).str.lower().map(keyword_to_category).fillna('Other')
```
